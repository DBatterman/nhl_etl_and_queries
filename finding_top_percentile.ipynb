```python
# Create a new dataframe that only has skaters in 5on5 situations

fiveOnFive_08_df = skaters_08_df.query("situation == '5on5'") 
```

```python
# View the new dataframe

fiveOnFive_08_df
```

```python
# Creating a function to figure out the 75th percentile of stats

def upper_percentile(x):
    return np.percentile(x, 75)
```

```python
# Applying our percentile function to stats we're potentially interested in

fiveOnFive_08_df[['I_F_xGoals', 'OnIce_F_xGoals',
                'I_F_highDangerShots', 'OnIce_F_highDangerShots',
                'I_F_highDangerxGoals','OnIce_F_highDangerGoals', 
                'penalityMinutesDrawn', 'penaltiesDrawn']].apply(upper_percentile)
```
